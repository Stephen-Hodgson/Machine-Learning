```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cesb2data.txt", index_col=0)
```

```python

derivatives = data.diff(periods=1)
data.head()
```

```python
p = '0.00'
Tcol = p + '_kbar_T'
rcool = p + '_kbar_r_cool'
rwarm = p + '_kbar_r_warm'

plt.plot(data.loc[:,Tcol], derivatives.loc[:,rcool]/derivatives.loc[:,Tcol])
plt.plot(data.loc[:,Tcol], derivatives.loc[:,rwarm]/derivatives.loc[:,Tcol])
plt.xlim((0, 20))
```
