```python
import seaborn as sns
```

```python
playId = 104

sns.scatterplot(x=tacklesdata.loc[tacklesdata.playId == playId].x, y=tacklesdata.loc[tacklesdata.playId == playId].y, hue=tacklesdata.loc[tacklesdata.playId == playId].displayName)
sns.scatterplot(x=tacklesdata.loc[(tacklesdata.playId == playId) & (tacklesdata.displayName == 'football')].x, y=tacklesdata.loc[(tacklesdata.playId == playId) & (tacklesdata.displayName == 'football')].y, label = 'football', color='k')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.show()
```
